### Display the state-wise mean rainfall for all the months using a pivot table.

```python
df_rain.head()
```

```python
df_rain_q4 = df_rain.drop(columns = ['DISTRICT','ANNUAL'])
#df_rain_q4
df_rain_q4 = df_rain_q4.melt(id_vars=['STATE_UT_NAME'], var_name='Month', value_name='Rainfall')
```

```python
df_rain_q4.groupby(by = ['STATE_UT_NAME','Month'])['Rainfall'].mean()
```
